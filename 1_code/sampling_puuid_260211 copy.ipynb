```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

def perform_stratified_sampling(sample_fraction, input_file, output_file, min_per_group=1, stratify_col='tier', random_state=42):
    """
    ê³„ì¸µì  ìƒ˜í”Œë§ ìˆ˜í–‰ (For Loop & Concat ë°©ì‹ - ë°ì´í„° ìœ ì‹¤ ë°©ì§€).
    """
    print('ğŸ“‚ ë°ì´í„° ë¡œë”© ì¤‘...')
    df = pd.read_csv(input_file)

    if stratify_col not in df.columns:
        raise ValueError(f"'{stratify_col}' ì»¬ëŸ¼ì´ ì…ë ¥ íŒŒì¼ì— ì—†ìŠµë‹ˆë‹¤.")

    # ê²°ê³¼ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸ ìƒì„±
    sampled_list = []

    # groupby ê°ì²´ë¥¼ ì§ì ‘ ìˆœíšŒ (ì´ ë°©ì‹ì´ ê°€ì¥ ì•ˆì „í•©ë‹ˆë‹¤)
    # group_nameì€ 'DIAMOND', 'MASTER' ê°™ì€ í‹°ì–´ ì´ë¦„, group_dfëŠ” í•´ë‹¹ í‹°ì–´ì˜ ë°ì´í„°í”„ë ˆì„
    for group_name, group_df in df.groupby(stratify_col):
        
        grp_size = len(group_df)
        
        # ìƒ˜í”Œ ê°œìˆ˜ ê³„ì‚°
        n = max(0, int(round(grp_size * sample_fraction)))
        
        # ìµœì†Œ ìƒ˜í”Œ ìˆ˜ ë³´ì¥ ë¡œì§
        if n < min_per_group and grp_size > 0:
            n = min(min_per_group, grp_size)
            
        # ìƒ˜í”Œë§ ìˆ˜í–‰
        if n > 0:
            sampled_group = group_df.sample(n=n, random_state=random_state)
            sampled_list.append(sampled_group)

    # ë¦¬ìŠ¤íŠ¸ì— ë‹´ê¸´ ê·¸ë£¹ë“¤ì„ í•˜ë‚˜ë¡œ ë³‘í•©
    if sampled_list:
        sampled_df = pd.concat(sampled_list, ignore_index=True)
    else:
        sampled_df = pd.DataFrame(columns=df.columns) # ë¹ˆ ê²°ê³¼ ì²˜ë¦¬

    # ê²°ê³¼ ì¶œë ¥
    print('\nâœ… ìƒ˜í”Œë§ ì™„ë£Œ!')
    print(f"ì „ì²´ ë°ì´í„° ìˆ˜: {len(df):,} ëª…")
    print(f"ìƒ˜í”Œë§ ë°ì´í„° ìˆ˜: {len(sampled_df):,} ëª…")
    
    # ë¹„ì¤‘ í™•ì¸ (ì´ì œ KeyErrorê°€ ë‚˜ì§€ ì•ŠìŠµë‹ˆë‹¤)
    if len(sampled_df) > 0:
        print(f"ìƒ˜í”Œ ë‚´ {stratify_col} ë¹„ì¤‘:\n{(sampled_df[stratify_col].value_counts(normalize=True) * 100).round(2)}%")

    # ì €ì¥
    sampled_df.to_csv(output_file, index=False, encoding='utf-8')
    print(f"ğŸ’¾ ìƒ˜í”Œ ë°ì´í„°ê°€ {output_file}ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
# 1. ê²½ë¡œ ì„¤ì •
file_path = r'D:\ë°”íƒ•í™”ë©´\ë°ì´í„° ë¶„ì„\í”„ë¡œì íŠ¸\04 ìµœì¢… í”„ë¡œì íŠ¸\Project_Sparta\0_data\csv ì •ë¦¬ í•„ìš”\br1'
name = 'br1_puuids_high_tiers'
sampling = 0.5

# 2. ì‹¤í–‰
perform_stratified_sampling(
    sample_fraction = sampling / 100,
    input_file = f'{file_path}\\{name}.csv', 
    output_file = f'{file_path}\\{sampling}%_{name}.csv',
    stratify_col = 'tier'
)
```
